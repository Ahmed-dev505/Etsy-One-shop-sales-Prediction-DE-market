```python
#importing libraries used for web scrapping
from bs4 import BeautifulSoup
import requests   
import pandas as pd
import time

def get_sold(url):


  """ fuction that returns a list of all items sold in a shop 
      args: url -> shop's url
      return: list
  """
  
  sold = [] ##empty list
  
  for i in range(571): #(they were 991)
    try:
      page = requests.get(url + str(i+1))
      soup = BeautifulSoup(page.text, 'html.parser')
      ##finding the content div
      content_div = soup.find_all("div", class_ = "content")[1] ##div containing content(link + title) as a part of "a" element 
      content_links = content_div.find_all("a",class_="listing-link")
    except:
      pass  
    for i in content_links:
      sold.append(i['href'])## list of items sold
    content_links = []
        

  return sold
  
```

```python
#experimental code for a proxy rotator

pip3 install bs4 requests stem

import requests
import random
from bs4 import BeautifulSoup as bs

def get_free_proxies():
    url = "https://free-proxy-list.net/"
    # get the HTTP response and construct soup object
    soup = bs(requests.get(url).content, "html.parser")
    proxies = []
    for row in soup.find("table", attrs={"id": "proxylisttable"}).find_all("tr")[1:]:
        tds = row.find_all("td")
        try:
            ip = tds[0].text.strip()
            port = tds[1].text.strip()
            host = f"{ip}:{port}"
            proxies.append(host)
        except IndexError:
            continue
    return proxies



#The below function accepts a list of proxies and creates a requests session that randomly selects one of the proxies passed:

def get_session(proxies):
    # construct an HTTP session
    session = requests.Session()
    # choose one random proxy
    proxy = random.choice(proxies)
    session.proxies = {"http": proxy, "https": proxy}
    return session


#Let's test this by making a request to a website that returns our IP address:


for i in range(5):
    s = get_session(proxies)
    try:
        print("Request page with IP:", s.get("http://icanhazip.com", timeout=1.5).text.strip())
    except Exception as e:
        continue
```

```python
def html_to_list(customer_name ,review_date,purchased_items,item_ratings):
  """
  function that takes in lists of html element and turns them into list of element content
  
  """
  for a,c,d,e in zip(customer_name,review_date,purchased_items,item_ratings):
    customer_names.append(a.text)
    #review_text_list.append(b.text)
    review_date_list.append(c.text)
    purchased_items_title_list.append(d.text)
    purchased_items_link_list.append(d['href'])
    item_ratings_list.append(e.find_next('span',class_='wt-screen-reader-only'))


def lists_to_dataframe(customer_names,review_date_list,purchased_items_title_list,purchased_items_link_list,item_ratings_list):
  """ simple function that make a dataframe our of lists"""
  allreviews = pd.DataFrame({'customer_name':customer_names,    
                             'review_date':review_date_list,
                             'item_title':purchased_items_title_list,
                             'item_link':purchased_items_link_list,
                             'item_rating':item_ratings_list 
                             
                             })
  return allreviews  
```

```python

from bs4 import BeautifulSoup
import requests   
import pandas as pd
import time
```

```python
#scrapping http://topcraftsellers.com/ for all top German shops

df = pd.DataFrame(columns=['shop_name','shop_link'])

for i in range(1,37):
  time.sleep(0.7)
  page = requests.get("http://topcraftsellers.com/top-etsy-sellers-in-germany/"+ str(i))
  soup = BeautifulSoup(page.text, 'html.parser')
  shop_table = soup.find("tbody") #finding the main table element. table element contain all shops details on page.

  shops  = [i for i in shop_table.find_all("a")] #list of a shop links on a certain page. link contains name and shop link.

  shop_names = [i['title'] for i in shops]
  shop_links = [i['href'] for i in shops]
  namesAndlinks = list(zip(shop_names,shop_links))
  df = df.append(pd.DataFrame(namesAndlinks,columns=['shop_name','shop_link']))

df.to_csv("germanShops1_topcraftsellers.csv")
```
